# Synapse comparison tool demo

Here we showcase a few examples of how the comparison tool would run. If you are not working with synapse entities, or you have transformed data inside a pipeline, you can call the reports generating function directly as well, and also view the reports directly

```python
# Import in our function for the following examples
from synapse_compare.compare_between_two_synapse_entities import run_compare
```

## Basic Synapse Table comparison
Here we compare between two synapse tables

```python
run_compare(
    syn_id_1="syn73836355",
    syn_id_2="syn73833080",
    version1="v1",
    version2="v2",
    compare_type="table",
    entity_name="maf_table",
    main_download_directory="comparison_output_tables",
    filter_on_version=False,
    save_to_synapse=False
)
```

## Basic Synapse File comparison
Here we compare between two different versions of a file

```python
compare_obj = run_compare(
    syn_id_1="syn52955202.17",
    syn_id_2="syn52955202.16",
    version1="releaseA",
    version2="releaseB",
    compare_type="file",
    entity_name="clinical_file",
    main_download_directory="comparison_output_files",
    filter_on_version=False,
)
```

## Using two dataframes directly

If you already have two dataframes read in available, you can run the generate the reports function directly. Here we generate some test clinical data.

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    "patient_id": ["P1", "P2", "P3", "P4"],
    "cohort": ["A", "A", "B", "C"],
    "age_at_diagnosis": [55, 63, 70, 45],
    "Tumor_Seq_Allele2": ["T", "CCACTGT", "G", "CCGCCGCCGCC"],
    "treatment_response": ["CR", "PR", "SD", "PD"]
})

df2 = pd.DataFrame({
    "patient_id": ["P1", "P2", "P3", "P5"],
    "cohort": ["A", "A", "B", "C"],
    "age_at_diagnosis": [55, 63, 71, 50],  # age difference for P3
    "Tumor_Seq_Allele2": ["T", "CCACTGT", np.nan, "CCGCCGCCGCC"],
    "treatment_response": ["CR", "PR", "PD", "PD"]
})
```

Now we use the generate_comparison_reports function directly

```python
from synapse_compare.compare_between_two_synapse_entities import generate_comparison_reports

reports = generate_comparison_reports(
    df1=df1,
    df2=df2,
    df1_name="df1_v1",
    df2_name="df2_v2",
    join_keys=["patient_id"]
)
```

### Viewing the datacompy report

```python
print(reports["comparison_report"].report())
```
