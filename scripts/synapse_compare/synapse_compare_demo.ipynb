## Basic Synapse Table comparison
Here we compare between two synapse tables

```python
run_compare(
    syn_id_1="syn12345678",
    syn_id_2="syn87654321",
    version1="v1",
    version2="v2",
    compare_type="table",
    entity_name="example_table",
    main_download_directory="comparison_output",
    filter_on_version=False,
    save_to_synapse=False
)
```

## Basic Synapse File comparison
Here we compare between two different versions of a file

```python
run_compare(
    syn_id_1="syn52955202.17",
    syn_id_2="syn52955202.16",
    version1="releaseA",
    version2="releaseB",
    compare_type="file",
    entity_name="clinical_file",
    main_download_directory="comparison_output_files",
    filter_on_version=False,
)
```

## Using custom pandas read_csv parameters
Here we add in custom pandas read_csv parameters for when we need to use beyond the default read_csv pandas settings to read our file in correctly

```python
run_compare(
    syn_id_1="syn12345678",
    syn_id_2="syn87654321",
    version1="v1",
    version2="v2",
    compare_type="file",
    entity_name="custom_csv_test",
    main_download_directory="comparison_output_csv",
    csv_kwargs={
        "sep": "\t",
        "dtype": {"sample_id": "string"},
        "low_memory": False,
    }
)
```
